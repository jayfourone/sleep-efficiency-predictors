```python
#import seaborn and matplotlib for fast/quick visuals
import seaborn as sns
import matplotlib.pyplot as plt

#create a histogram visualisation of sleep quality data
#kde = kernal density estimate curve
#bins = intervals
sns.histplot(df["Quality of Sleep"], kde = True, bins = 15)
plt.title("Distribution of Quality of Sleep Ratings")
plt.xlabel("Quality of sleep (1-10)")
```

```python
#compare target vs predictors
#target = quality of sleep
#predictors = other features/components 
#sns.boxplot(data= , x= ,y= '
genderPlot = sns.boxplot(x="Gender", y="Quality of Sleep", data=df)
```

```python
#use regplot in seaborn which plots scatterplot with regression line
ActivityPlot = sns.regplot(x = "Physical Activity Level", y = "Quality of Sleep", data = df)
```

```python
# ^^^
stressPlot = sns.regplot(x = "Stress Level", y = "Quality of Sleep", data = df)
```

```python
#correlation heatmap

corr = df.corr(method= 'pearson', numeric_only = True)
sns.heatmap(corr, annot = None, cmap = "coolwarm")
plt.title("Correlation Heatmap")
```
